```python
#import module
import folium
# crear un objeto # Latitud y Longitud
my_map = folium.Map(location=[-8.390805,-74.572451],
                    zoom_start=12)
my_map
```

```python
# agregar base mapa en folium
basemaps = {
    'Google Maps': folium.TileLayer(
        tiles = 'https://mt1.google.com/vt/lyrs=m&x={x}&y={y}&z={z}',
        attr = 'Google',
        name = 'Google Maps',
        overlay = True,
        control = True
    ),
    'Google Satellite': folium.TileLayer(
        tiles = 'https://mt1.google.com/vt/lyrs=s&x={x}&y={y}&z={z}',
        attr = 'Google',
        name = 'Google Satellite',
        overlay = True,
        control = True
    ),
    'Google Terrain': folium.TileLayer(
        tiles = 'https://mt1.google.com/vt/lyrs=p&x={x}&y={y}&z={z}',
        attr = 'Google',
        name = 'Google Terrain',
        overlay = True,
        control = True
    ),
    'Google Satellite Hybrid': folium.TileLayer(
        tiles = 'https://mt1.google.com/vt/lyrs=y&x={x}&y={y}&z={z}',
        attr = 'Google',
        name = 'Google Satellite',
        overlay = True,
        control = True
    ),
    'Esri Satellite': folium.TileLayer(
        tiles = 'https://server.arcgisonline.com/ArcGIS/rest/services/World_Imagery/MapServer/tile/{z}/{y}/{x}',
        attr = 'Esri',
        name = 'Esri Satellite',
        overlay = True,
        control = True
    )
}
```

```python
# Add custom basemaps
basemaps['Google Satellite'].add_to(my_map)
my_map
```

```python
excel = pd.read_excel("Excel/COORDENADAS.xlsx")
```

```python
excel.columns
```

```python
# Eliminar espacios en nombres campos
excel.columns  = excel.columns.str.strip()
```

```python
# Create point geometries
geometria = gpd.points_from_xy(excel['ESTE'],excel['NORTE'], crs=32718)
```

```python
datos_shp = gpd.GeoDataFrame(excel, geometry=geometria)
```

```python
datos_shp.head()
```

```python
len(datos_shp)
```

```python
# Reproyectar a Geografico
shp_GEO = datos_shp.to_crs(epsg=4326)
```

```python
Longitud = shp_GEO.geometry.x
Latitud = shp_GEO.geometry.y
```

```python
Latitud.mean()
```

```python
map1 = folium.Map(location=[Latitud.mean(),
                           Longitud.mean()],
                 zoom_start=12, 
                 control_scale=True)

basemaps['Google Satellite'].add_to(map1)
map1
```

```python
shp_GEO["Sexo"]
```

```python
# add marker one by one on the map
for i in range(0,len(shp_GEO)):
    folium.Marker(location=[Latitud[i],
                            Longitud[i]],
                   popup=shp_GEO["Sexo"][i]).add_to(map1)
map1
```
